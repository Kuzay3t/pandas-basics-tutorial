## **PANDAS**

Pandas is a popular framework used to prepare and handle data set. To use the libary, it has to be imported.

The two core objects in pandas which are:
1.   Dataframe: Handles data in a tabular form i.e rows and columns
2.   Series: A Series, by contrast, is a sequence of data values. If a DataFrame is a table, a Series is a list.



```python
# To import pandas into the notebook

import pandas as pd
```

Creating Dataframes

```python
# A dataframe is created using a .dataframe() method
pd.DataFrame({"A":[1,2,3,4,5], "B":[6,7,8,9,10]})
```

```python
# creating a dataframe
pd.DataFrame([[34,54], [45,98]] , columns = ['Age', 'Weight'], index = [' First Year', 'Second Year'])
```

Creating Series


```python
# Series are created using .Series method
pd.Series([1,2,3,4])
```

```python
pd.Series([1,2,3], index = ['Baby', 'Toddlers', 'Childrens'], name = 'Children')
```
